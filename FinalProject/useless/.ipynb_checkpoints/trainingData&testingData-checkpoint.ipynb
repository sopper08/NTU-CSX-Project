# Mark

## Drop OrderGroupCode & ECouponId

```python
df_mark.drop("OrderGroupCode", axis=1, inplace=True)
df_mark.head()
```

# Label

```python
df_label = pd.read_csv("data/proData/label.csv")
df_label.head()
```

## drop 0 次購買人

```python
import numpy as np
df_label.replace(0, np.NAN, inplace=True)
df_label.dropna(inplace=True)
df_label.head()
```

## 購買次數 v.s. 人數

```python
order_list = df_label["OrderGroupCode"]
temp_dict = { 1 : 0 }
for i in order_list:
#     print(i)
    try:
        num = temp_dict[int(i)]
    except:
        num=0
    num = num + 1
    temp_dict[int(i)] = num
temp_dict
```

```python
df_num = pd.DataFrame()
df_num["numofOrder"] = temp_dict.keys()
df_num["num"] = temp_dict.values()
df_num.sort_values("numofOrder", inplace=True)
df_num.head(30)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(df_num['numofOrder'], df_num['num'], 'o')
plt.xlim(0, 100)
```

```python
plt.plot(df_num['numofOrder'], df_num['num'], 'o')
plt.xlim(8, 20)
plt.ylim(0, 10000)
```
