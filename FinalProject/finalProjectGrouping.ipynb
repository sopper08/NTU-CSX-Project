先確定開始及結束時間的資料格式

````python
df_Promo_1.info()
````

發現為object格式，須轉換成datetime

````python
df_Promo_1['PromotionStartDateTime'] = pd.to_datetime(df_Promo_1['PromotionStartDateTime'], format='%Y-%m-%d %H:%M:%S.%f')
df_Promo_1['PromotionEndDateTime'] = pd.to_datetime(df_Promo_1['PromotionEndDateTime'], format='%Y-%m-%d %H:%M:%S.%f')
df_Promo_1.info()
````

產生一個DataFrame，index為日期，欄位是活動內容
所以需要把開始到結束日期對應到新的DataFrame

````python
df_DCType0 = df_Promo_1[df_Promo_1['DiscountType'].isnull()]
df_DCType0.head()
````

````python
days = pd.date_range('2015-05-06 23:59:59', '2018-01-08', freq='1D')
df_DCT = pd.DataFrame({'date': days})
df_DCT.head()
````

````python
dictOfDCType0 = {}
for i in df_DCType0.index:
#     print(df_DCType0['PromotionStartDateTime'][i]);
    for j in days:
#         print(df_DCType0['PromotionStartDateTime'][i]);
#         print(j)
#         print("---")
        if df_DCType0['PromotionStartDateTime'][i]<j:
            if j>df_DCType0['PromotionEndDateTime'][i]:
                break
            dictOfDCType0[j] = 0
dictOfDCType0
````

接著將type0加入dataFrame
* dict如何加入dataframe？
    * 以dict新增一個欄位，將key對應至某個欄位，並將value放入新的欄位。
    參考自：http://cmdlinetips.com/2018/01/how-to-add-a-new-column-to-using-a-dictionary-in-pandas-data-frame/
```
df['新的欄位名稱'] = df['對應的欄位名稱'].map(放置dict)
```

````python
df_DCT['type0'] = df_DCT['date'].map(dictOfDCType0)
df_DCT.head()
````

剩下的type也是用同樣的方法map進去
* NAN >> type0 
* 任選優惠價 >> type1
* 滿件打折 >> type2
* 滿件折現 >> type3
* 滿額打折 >> type4
* 滿額折現 >> type5
* 滿額贈 >> type6

````python
dictOfDCType = {}
# type1
df_DCType1 = df_Promo_1[df_Promo_1['DiscountType']=='任選優惠價']
for i in df_DCType1.index:
    for j in days:
        if df_DCType1['PromotionStartDateTime'][i]<j:
            if j>df_DCType1['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 1
df_DCT['type1'] = df_DCT['date'].map(dictOfDCType)

# type2
dictOfDCType = {}
df_DCType2 = df_Promo_1[df_Promo_1['DiscountType']=='滿件打折']
for i in df_DCType2.index:
    for j in days:
        if df_DCType2['PromotionStartDateTime'][i]<j:
            if j>df_DCType2['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 2
df_DCT['type2'] = df_DCT['date'].map(dictOfDCType)

# type3
dictOfDCType = {}
df_DCType3 = df_Promo_1[df_Promo_1['DiscountType']=='滿件折現']
for i in df_DCType3.index:
    for j in days:
        if df_DCType3['PromotionStartDateTime'][i]<j:
            if j>df_DCType3['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 3
df_DCT['type3'] = df_DCT['date'].map(dictOfDCType)  

# type4
dictOfDCType = {}
df_DCType4 = df_Promo_1[df_Promo_1['DiscountType']=='滿額打折']
for i in df_DCType4.index:
    for j in days:
        if df_DCType4['PromotionStartDateTime'][i]<j:
            if j>df_DCType4['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 4
df_DCT['type4'] = df_DCT['date'].map(dictOfDCType)

# type5
dictOfDCType = {}
df_DCType5 = df_Promo_1[df_Promo_1['DiscountType']=='滿額折現']
for i in df_DCType5.index:
    for j in days:
        if df_DCType5['PromotionStartDateTime'][i]<j:
            if j>df_DCType5['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 5
df_DCT['type5'] = df_DCT['date'].map(dictOfDCType)

# type6
dictOfDCType = {}
df_DCType6 = df_Promo_1[df_Promo_1['DiscountType']=='滿額贈']
for i in df_DCType6.index:
    for j in days:
        if df_DCType6['PromotionStartDateTime'][i]<j:
            if j>df_DCType6['PromotionEndDateTime'][i]:
                break
            dictOfDCType[j] = 6
df_DCT['type6'] = df_DCT['date'].map(dictOfDCType)
df_DCT.to_csv('DCT.csv')
````

先將date換成index，把0換成nan，為了讓繪圖不要繪製出0

````python
import numpy as np
df_DCT_dateindex = pd.DataFrame(df_DCT.groupby('date').sum())
cols = ["type0", "type1", "type2", "type3", "type4", "type5", "type6"]
df_DCT_dateindex[cols] = df_DCT_dateindex[cols].replace({0:np.nan})
df_DCT_dateindex.head()
````
