```python
import pandas as pd
df_ori = pd.read_csv("./data/proData/mergeData.csv")
df_ori.head()
```

```python
df_seg = df_ori.copy()
```

## 建立 Label 對照表

```python
# lab: label
df_lab = df_seg[["MemberId", "OrderGroupCode"]].copy()
df_lab["OrderGroupCode"][df_lab["OrderGroupCode"].notnull()] = 1
data = df_lab.groupby("MemberId")["OrderGroupCode"].sum()

df_lab = pd.DataFrame(data).reset_index()
df_lab.head()
```

```python
df_lab.sort_values("OrderGroupCode", ascending=False)
```

```python
# 0次購買紀錄
label_list = []
temp_list = df_lab[df_lab["OrderGroupCode"]==0]['MemberId'].tolist()
label_list.append(temp_list)

# 1~277次購買紀錄
for i in range(0, 277):
    temp_list = []
    temp_list = df_lab[df_lab["OrderGroupCode"]>i]['MemberId'].tolist()
    label_list.append(temp_list)
```

## 未購買的 Member

```python
df_seg_0 = df_seg[df_seg["OrderGroupCode"].isnull()] 
df_seg_0.to_csv("./data/segData/df_seg_0.csv", index=False)
df_seg_0.head()
```

### 去除掉未購買的 Member

```python
df_seg.dropna(inplace=True)
df_seg.head()
```

## 上編號 第 n 筆訂單

```python
num_list=[]
num = 1
pre_memId = 0
for i in df_seg.index:
#     print(i, end='\r', flush=True)
#     print(df_seg.loc[i, "MemberId"])
    memId = df_seg.loc[i, "MemberId"]
    if memId == pre_memId:
        num = num + 1
        num_list.append(num)
    else:
        pre_memId = memId
        num = 1
        num_list.append(num)
```

```python
df_seg["num"]=num_list
df_seg.head(30)
```

```python
def seg(a, df):
    df_S = df[df["num"]<=a]
    df_S.to_csv("./data/segData/df_seg_{0}.csv".format(a), index=False)
```

```python
for i in range(1, 10):
    print(i)
    seg(i, df_seg)
    
#     = df_seg[df_seg["OrderGroupCode"]<=i].head()
```
