將蕭敬騰專輯的11首歌分別作字詞分析

```python
import jieba.analyse

# 建立一個taglist儲存每首個的前15 tags
tagList = []

for i in range(1, 12):
    with open("data/test%d.txt"%i, "rb") as f:
        text = f.read()
        tags = jieba.analyse.extract_tags(text.decode(), 15)
        tagList = tagList + tags;
    f.close()
print(tagList)
```

```python
text = ' '.join(tagList)
tags = jieba.analyse.extract_tags(text, 15)
print(tags)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud


stopwords = {}.fromkeys(["我的","我們","我們的","還是","什麼"])

wc = WordCloud(font_path="NotoSerifCJKtc-Black.otf", #設置字體
               background_color="white", #背景顏色
               max_words=2000, #文字雲顯示最大詞數
               stopwords=stopwords) #停用字詞

wc.generate(text)

# 視覺化
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```
