```python
import requests
import json
import pandas as pd #引用pandas函式庫並以pd簡稱
#記得在給URL時要把API裡的callback刪掉, 回傳資料才符合Json格式
res = requests.get("https://s.taobao.com/search?data-key=s&data-value=44&ajax=true&_ksTS=1520992126043_959&q=%E4%BF%9D%E6%BA%AB%E6%9D%AF&type=p&tmhkh5=&spm=a21wu.241046-tw.a2227oh.d100&from=sea_1_searchbutton&catId=100&bcoffset=4&ntoffset=0&p4ppushleft=1%2C48")
data = json.loads(res.text)
print(data)
```
