```python
import pandas as pd
df_ori = pd.read_csv('data/train.csv')
df_ori.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

plt.figure()

cm = plt.cm.get_cmap('jet')
area = np.pi * 8
sc = plt.scatter(
    df_ori['sex'], 
    df_ori['age'], 
    s=area, 
    c=df_ori['survived'],
    cmap=cm, 
    alpha=0.5)
plt.colorbar(sc)
plt.show()
```

male死亡率較高，male年輕人死亡率較低。

```python
plt.figure()

cm = plt.cm.get_cmap('jet')
area = np.pi * 8
sc = plt.scatter(
    df_ori['age'], 
    df_ori['pclass'], 
    s=area,
    c=df_ori['survived'],
    cmap=cm, 
    alpha=0.5)
plt.colorbar(sc)
plt.show()
```

Pclass存活率：1st > 2nd > 3rd

```python
plt.figure()

cm = plt.cm.get_cmap('jet')
area = np.pi * 8
sc = plt.scatter(
    df_ori['pclass'], 
    df_ori['fare'], 
    s=area,
    c=df_ori['survived'],
    cmap=cm, 
    alpha=0.5)
plt.colorbar(sc)
plt.show()
```
