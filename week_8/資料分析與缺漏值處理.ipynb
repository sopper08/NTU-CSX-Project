### 欄位處理：
* name: 不使用。
* sex: female:0, male:1.
* age: 缺漏值使用平均年齡。(也許也可以先以機器學習估計出年齡)
* cabin: change to [A, B, C, D, E, F]欄位，其中 nan : [0, 0, 0, 0, 0, 0]。  
* sibsp: change to [sibsp1, sibsp2, sibsp3, sibsp4, sibsp5], 其中 0 : [0, 0, 0, 0, 0]。  
* parch: change to [parch1, parch2, parch3, parch4, parch5, parch6], 其中 0 : [0, 0, 0, 0, 0, 0]。
* ticket: 不使用。  
* fare: 不處理直接使用。  
* embarked: change to [C, Q, S], 其中 nan: [0, 0, 0]。  
* boat: change to [A, B, C, D, boat1, boat2, boat3, … boat12]。
* body: 不使用。
* home.dest: 不使用。
