# webCrawler_CWDDailyAvgTempOf466920
爬取氣象局2017年466920測站的平均溫度

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import lxml
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

monthList=[31,28,31,30,31,30,31,31,30,31,30,31]
tempList = []

for i in range(1,12+1):
    url = 'http://e-service.cwb.gov.tw/HistoryDataQuery/MonthDataController.do?command=viewMain&station=466920&stname=%25E8%2587%25BA%25E5%258C%2597&datepicker=2017-' + '%02d'%((i))
#     print(url)
    temp=pd.read_html(url)
#     print(list[i])
    for j in range(1,monthList[i-1]+1):
        tempList.append(temp[1][7][j+1])
```

```python
dailyIndex = pd.date_range('1/1/2017 00:00', periods=365, freq='D')
s_DailyAvgTemp = pd.DataFrame({
    'new_DateId': dailyIndex,
    'DailyAvgTemp': tempList})
s_DailyAvgTemp
```

```python
s_DailyAvgTemp.info()
```

需更改DailyAvgTemp的型態為float

```python
s_DailyAvgTemp['DailyAvgTemp'] = s_DailyAvgTemp['DailyAvgTemp'].astype(float)
s_DailyAvgTemp.info()
```

確定正確後儲存成csv檔

```python
s_DailyAvgTemp.to_csv('dailyAvgTemp2017.csv', index=True, header=True)
```

```python
data = s_DailyAvgTemp.groupby('new_DateId')['DailyAvgTemp'].sum()
%matplotlib inline 
data.plot()
```
