```python
import pandas as pd
import matplotlib.pyplot as plt # 繪圖用

df = pd.read_csv('Ntu_Orders.csv')
df['new_DateId'] = pd.to_datetime(df['DateId'].astype(str), format='%Y%m%d')
df_App = df[df['TrackSourceTypeDef'] != 'Web']
df_iOSApp = df_App[df_App['TrackSourceTypeDef'] == 'iOSApp']
df_AndroidApp = df_App[df_App['TrackSourceTypeDef'] == 'AndroidApp']
df_Web = df[df['TrackSourceTypeDef'] == 'Web']
```

### 總訂單量的比較
訂單量: 依購物車編號(TradesOrderGroupCode)決定, 同一編號代表同一訂單, 不重複計算.

```python
orderQuantityFromApp = df_App.groupby("TradesOrderGroupCode").size().shape[0]
orderQuantityFromiOSApp = df_iOSApp.groupby("TradesOrderGroupCode").size().shape[0]
orderQuantityFromAndroidApp = df_AndroidApp.groupby("TradesOrderGroupCode").size().shape[0]
orderQuantityFromWeb = df_Web.groupby("TradesOrderGroupCode").size().shape[0]

print("來自APP的訂單量: ", orderQuantityFromApp)
print("    來自iOSApp的訂單量:     ", orderQuantityFromiOSApp)
print("    來自AndroidApp的訂單量: ", orderQuantityFromAndroidApp)
print("來自Web的訂單量: ", orderQuantityFromWeb)
```

```python
labels = ["App", "Web"]
sizes = [orderQuantityFromApp, orderQuantityFromWeb]
explode = (0, 0.1)
colors = ['yellow', 'green']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 90, radius = 1.2)
ax1.axis('equal')

plt.show()
```

發現若不分時間, 算App與Web的佔比, 似乎沒有太大意義, 故以下圖表以時間(new_DateId)當橫軸.

### 首先探討來自Web的訂單量變化

```python
s_WebVSOrderQuantity = pd.Series(df_Web.groupby("new_DateId").TradesOrderGroupCode.nunique())
%matplotlib inline
s_WebVSOrderQuantity.plot()
```

觀察日期與訂單的關係後, 發現以下幾點:
1. 日期是從2017-01-01~2018-01-09
2. 若以每日來觀察訂單量的關係, 區間太過狹小, 故改成每個月!

```python
%matplotlib inline
s_WebVSOrderQuantity.resample('M').sum().plot(kind='bar')
```

從圖中發現訂單在3.4月達到高峰, 接著下降, 又到了11月回到高峰, 猜測是因為換季潮!
接著去掉取消及失敗的訂單, 再做一次做圖分析!

```python
df_Web_RealOQ = pd.DataFrame(df_Web[(df_Web['SalesOrderSlaveStatusDef'] != 'Cancel') & (df_Web['SalesOrderSlaveStatusDef'] != 'Fail')])
s_Web_RealOQ =pd.Series(df_Web_RealOQ.groupby("new_DateId").TradesOrderGroupCode.nunique())
%matplotlib inline
s_Web_RealOQ.resample('M').sum().plot(kind='bar')
```

趨勢差不多, 看來取消及失敗的訂單並不會因為月份而有改變

### 接著探討來自App的訂單量變化
一樣分析以下兩者:
1. 訂單量與時間(月)
2. 去掉失敗與取消的訂單量與時間(月)

```python
s_AppVSOrderQuantity = pd.Series(df_App.groupby("new_DateId").TradesOrderGroupCode.nunique())
%matplotlib inline
s_AppVSOrderQuantity.resample('M').sum().plot(kind='bar')
```

滿有趣的, 在APP端高峰一樣是在3.4月及11.12月, 但App在11.12月較高, 不同於Web.
以下是兩種猜測:
1. APP推廣有成
2. 天氣冷不想用電腦

試著以平均溫度與訂單量的關係, 看看第二點的猜測對不對!
