```python
# 新增一個new_DataId變數, 沿用DataId的值並更改格式
df['new_DateId'] = pd.to_datetime(df['DateId'].astype(str), format='%Y%m%d')
df.info() #檢查有無出現新的變數
df.head() #檢查有無出現日期格式
```

```python
df['Quantity'].head() #取特定欄位的前五筆資料
```

```python
df.loc[1000:1005, ['new_DateId', 'Quantity']] #顯示特定幾筆/欄位的資料
```

```python
df[df['Quantity'] > 1] #提供條件, 篩選符合的資料
```

```python
data = df.groupby(by='new_DateId')['Quantity'].sum() #將特定資料作為index算特定欄位的總和
data.head(10)
```
